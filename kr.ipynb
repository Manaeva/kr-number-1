```python
x = int(input('Введите X:'))
s = 0
for i in range(11):
    s = s + 1/x**i
print(f'S = {s}')
```

```python
n = 30
norm  = int(input('Введите заданный норматив:'))
res = []
while n != 0:
    b1 = int(input('Введите результат:'))
    n -= 1
    res.append(b1)
k = 0
for i in res:
    if i >= norm:
        k += 1
print(f'{k} учеников выполнили норматив!')
```

```python
m = [int(i) for i in input('Введите числа в массив через пробел:').split()]
print(f'Начальный массив {m}')
#find max
mx = -10**8
for i in m:
    mx = max(i,mx)
#append i
ans = []
for i in range(len(m)):
    if m[i] == mx:
        ans.append(i)
print(f'Индексы максимальных элементов:{ans}')
```
